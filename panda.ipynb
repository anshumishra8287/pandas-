# Data handilling
## Handling missing values is a common task when working with data in pandas. Missing values in pandas are usually represented as NaN (Not a Number). Here’s a comprehensive guide to dealing with them:



````python
# data handling :
data= {
    "name":['ram',None,'ghanshyam','aditi','dhanu','rituraj','infinite','jagdish'],
    "Age" :[28,None,22,30,29,40,25,32],
    "salary":[50000,None,45000,52000,49000,70000,48000,58000],
    "Performance Score":[85,None,78,92,88,95,80,89]
}
df=pd.DataFrame(data)
print(df)
print(df.isnull())# True where values are missing
print(df.isnull().sum())# Count of missing values per column
````

````python
#droping a missing value :
data= {
    "name":['ram',None,'ghanshyam','aditi','dhanu','rituraj','infinite','jagdish'],
    "Age" :[28,None,22,30,29,40,25,32],
    "salary":[50000,None,45000,52000,49000,70000,48000,58000],
    "Performance Score":[85,None,78,92,88,95,80,89]
}
df=pd.DataFrame(data)
print(df)
#Drop rows with any missing values:
df.dropna(inplace=True)
print(df)
````

````python
#droping a missing value :
data= {
    "name":['ram',None,'ghanshyam','aditi','dhanu','rituraj','infinite','jagdish'],
    "Age" :[28,None,22,30,29,40,25,32],
    "salary":[50000,None,45000,52000,49000,70000,48000,58000],
    "Performance Score":[85,None,78,92,88,95,80,89]
}
df=pd.DataFrame(data)
print(df)

#Drop rows where all values are missing:
df.dropna(how='all', inplace=True)

print(df)
````

````python
# fill a missing value with constant :
import pandas as pd
data= {
    "name":['ram',None,'ghanshyam','aditi','dhanu','rituraj','infinite','jagdish'],
    "Age" :[28,None,22,30,29,40,25,32],
    "salary":[50000,None,45000,52000,49000,70000,48000,58000],
    "Performance Score":[85,None,78,92,88,95,80,89]
}
df=pd.DataFrame(data)
print(df)
# replace or none value in age with unknown:
df['Age']=df['Age'].fillna('unknown')
print(df)
# replace  all value where none ar nan are present with 0:
df=df.fillna(0)
print (df)
````

````python
# fill a missing value with mean ,mode,median :
import pandas as pd
data= {
    "name":['ram',None,'ghanshyam','aditi','dhanu','rituraj','infinite','jagdish'],
    "Age" :[28,None,22,30,29,40,25,32],
    "salary":[50000,None,45000,52000,49000,70000,48000,58000],
    "Performance Score":[85,None,78,92,88,95,80,89]
}
df=pd.DataFrame(data)
print(df)
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
````

In pandas, **concatenation** means combining multiple `DataFrame` or `Series` objects together using `pd.concat()`.

## Basic Syntax

```python
import pandas as pd

pd.concat(objs, axis=0)
```

* `objs` → list of DataFrames/Series
* `axis=0` → combine row-wise (default)
* `axis=1` → combine column-wise

---

# 1. Row-wise Concatenation (`axis=0`)

```python
import pandas as pd

df1 = pd.DataFrame({
    'Name': ['Aman', 'Riya'],
    'Age': [21, 22]
})

df2 = pd.DataFrame({
    'Name': ['Karan', 'Neha'],
    'Age': [23, 24]
})

result = pd.concat([df1, df2])

print(result)
```

### Output

```python
    Name  Age
0   Aman   21
1   Riya   22
0  Karan   23
1   Neha   24
```

---

# 2. Reset Index

```python
result = pd.concat([df1, df2], ignore_index=True)
```

### Output

```python
    Name  Age
0   Aman   21
1   Riya   22
2  Karan   23
3   Neha   24
```

---

# 3. Column-wise Concatenation (`axis=1`)

```python
df1 = pd.DataFrame({
    'Name': ['Aman', 'Riya']
})

df2 = pd.DataFrame({
    'Age': [21, 22]
})

result = pd.concat([df1, df2], axis=1)

print(result)
```

### Output

```python
   Name  Age
0  Aman   21
1  Riya   22
```

---

# 4. Concatenate Multiple DataFrames

```python
result = pd.concat([df1, df2, df3])
```

---

# 5. Join Types

```python
pd.concat([df1, df2], axis=1, join='inner')
```

* `outer` → keep all columns/rows (default)
* `inner` → keep only common indexes

---

# Difference Between `concat()` and `merge()`

| concat()                    | merge()             |
| --------------------------- | ------------------- |
| Combines along rows/columns | Combines using keys |
| Simpler                     | SQL-like join       |
| Uses axis                   | Uses common column  |

---
